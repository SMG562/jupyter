# 2. clean data 

```python
def clean_data(fileName):
    
    cleanData = []
    wordDict = []
    
    inputFile = open("training.txt", "r")
    
    for line in inputFile: 
        line_part = line.split('#')
        print(line_part)

        label = line_part[0]
        print(label)

        words = line_part[1].split(" ")
        print(words)


        cleanData.append((words,label)) 

        for word in words:
            wordDict.append(word)

    return(cleanData, wordDict)
```

```python
clean_data('training.txt')
```
