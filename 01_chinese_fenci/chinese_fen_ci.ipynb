```python
import jieba

jieba.del_word('作者')
jieba.del_word('作 者')
```

```python
# read file into a variable. 
with open("three_hundred_tang_poems.txt") as f:
    tang_peoms_text= f.read()
```

```python
type(tang_peoms_text)
```

```python
print(tang_peoms_text)
```

```python
seg_list = jieba.cut(tang_peoms_text, cut_all=False) # precise 模式

print(seg_list)
```

```python
# use space ' ' to seperate the words. 


seg_list_space = ' '.join(seg_list)

print(seg_list_space) 
```

```python
# install the wordcloud lib using pip or conda 
# $ pip install wordcloud
# or 
# $ conda install -c conda-forge wordcloud


# import the word cloud lib.
from wordcloud import WordCloud

tang_peom_word_cloud = WordCloud(font_path="simsun.ttf").generate(seg_list_space)

```

```python


import matplotlib.pyplot as plt

plt.imshow(tang_peom_word_cloud, interpolation='bilinear')

plt.axis('off')

```

```python
import jieba.analyse
#jieba.analyse.set_stop_words('/pyth/of/stop/words')

tags = jieba.analyse.extract_tags(seg_list_space, topK=20,withWeight = True)

print('type of tags:', type(tags))

# print(",".join(tags))

for v,n in tags:
    print(v + '\t' + str(int(n*10000)))
          
          
```

```python
print('-'*40)
print(' TextRank')
print('-'*40)


for x, w in jieba.analyse.textrank(seg_list_space, withWeight=True):
    print('%s %s' % (x, w))
```

```python
print('-'*40)
print(' TextRank')
print('-'*40)


for x, w in jieba.analyse.textrank(seg_list_space, topK=20, withWeight=False):
    print('%s %s' % (x, w))
```
