```python
# read the file into a variable, here the variable is 'hamlet_text'

with open('hamlet.txt') as f:
    hamlet_text = f.read()
```

```python
# output the variable

hamlet_text
```

```python

# you need to import the library wordcloud
# import means re-use the function that others has built aldready.

# from ... import ... means we only import part of the function, like we buy only one piece of pizza, rather than the whole pizza. 

# before import, you need to install the package by running
# $ conda install -c https://conda.anaconda.org/conda-forge wordcloud


from wordcloud import WordCloud
hamlet_wordcloud = WordCloud().generate(hamlet_text)
```

```python
# draw the picture
# For visualization, matplotlib is a basic library that enables many other libraries to run and plot on its base 
import matplotlib.pyplot as plt
plt.imshow(hamlet_wordcloud, interpolation='bilinear')
plt.axis('off')

```
